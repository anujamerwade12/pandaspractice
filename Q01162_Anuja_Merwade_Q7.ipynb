```python
import pandas as pd
```

```python
df = pd.read_csv('football.csv')
```

```python
df.dtypes
```

```python
df['Value'] = df['Value'].astype(str)
```

```python
df.head()
```

```python
df['Value_without_euro'] = df['Value'].str.replace('€','')
```

```python
df['Value_without_euro'] = df['Value_without_euro'].astype(str)
```

```python
def func_Memory(line):
    if line[-1] == "M":
        return float(line.replace('M',''))*1000000
    elif line[-1] == "K":
        return float(line.replace('K',''))*1000
    else: return line
df['Value_converted'] = df['Value_without_euro'].apply(func_Memory)
#print(func_Memory(df['Value_without_euro']))
print(df['Value_converted'])
```
